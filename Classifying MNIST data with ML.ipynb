```python
from numpy import *
from pandas import *
from matplotlib.pyplot import *
from seaborn import *
from datetime import *
from sklearn.decomposition import PCA

import matplotlib
import time
```

```python
# Settings
# data = 32*32
trainPath = 'C:/Users/Affinity/Documents/digits/trainingDigits/'
testPath = 'C:/Users/Affinity/Documents/digits/testDigits/'
numDigits_train = {0: 189, 1: 198, 2: 195, 3: 199, 4: 186, 5: 187, 6: 195, 7: 201, 8: 180, 9: 204}
numDigits_test = {0: 87, 1: 97, 2: 92, 3: 85, 4: 114, 5: 108, 6: 87, 7: 96, 8: 91, 9: 89}
nTrain=1934; nTest=946
```

## 1.1. Reading in txt data & preprocessing

#### Preprocessing train data

```python
# preprocessing train data
ct=-1
trainData=zeros((nTrain,32**2))
trainLabel=zeros((nTrain,1))
for d in arange(10):
    currN = numDigits_train[d]
    for n in arange(currN):
        ct+=1
        temp=read_csv(trainPath+str(d)+ '_' +str(n)+ '.txt', header=None)
        temp1=[]
        for a in arange(32):
            temp1.extend(temp[0][a])
        trainData[ct]=temp1
        trainLabel[ct] = str(d)
print(trainData.shape)
print(trainLabel.shape)
```

#### Preprocessing test data

```python
# preprocessing test data
ct=-1
testData=zeros((nTest,32**2))
testLabel=zeros((nTest,1))
for d in arange(10):
    currN = numDigits_test[d]
    for n in arange(currN):
        ct+=1
        temp=read_csv(testPath+str(d)+ '_' +str(n)+ '.txt', header=None)
        temp1=[]
        for a in arange(32):
            temp1.extend(temp[0][a])
        testData[ct]=temp1
        testLabel[ct] = str(d)
print(testData.shape)
print(testLabel.shape)
```

### 1.3.1. Applying SVM

```python
from sklearn.svm import SVC
from sklearn.metrics import *
```

#### Hyperparameter Tuning

```python
# Merging the train data with its label and then shuffling 

temp_train=concatenate((pca_train, trainLabel),axis=1)
temp_train=DataFrame(temp_train)
temp_train.head()
```

```python
temp_train1=temp_train.reindex(np.random.permutation(temp_train.index))
```

```python
temp_train1.head()
```

```python
# Checking whether sorted correctly
unique(temp_train.iloc[int(temp_train1.iloc[0].name),:]==temp_train1.iloc[0,:])
```

```python
# Splitting the data and label again
train1 = temp_train1.iloc[:1000,:-2]
train2 = temp_train1.iloc[1001:,:-2]
label1 = temp_train1.iloc[:1000,-1]
label2 = temp_train1.iloc[1001:,-1]

print(train1.shape, train2.shape, label1.shape, label2.shape)
```

```python
# Set the parameters by cross-validation:
# https://www.kaggle.com/udaysa/svm-with-scikit-learn-svm-with-parameter-tuning
```

```python
# Grid search
best_score=0
for gamma in [0.001, 0.01, 0.1, 1, 10, 100]:
    for C in [0.001, 0.01, 0.1, 1, 10, 100]:
        # 매개변수의 각 조합에 대해 SVC를 훈련시킵니다.
        svm = SVC(gamma=gamma, C=C)
        svm.fit(train1,label1)
        # 테스트 세트로 SVC를 평가합니다
        score = svm.score(train2,label2)
        # 점수가 더 높으면 매개변수와 함께 기록
        if score>best_score:
            best_score=score
            best_parameters={'C':C, 'gamma':gamma}
print("최고 점수: {:.2f}".format(best_score))
print("최적 매개변수:",best_parameters)
```

#### SVM Classification

```python
svc=SVC(**best_parameters)
```

```python
svc.fit(pca_train,trainLabel)
```

```python
a=svc.predict(pca_test)
a.shape
```

```python
info(testLabel)
```

```python
errors=0
accuracyTable=zeros(nTest)
for t in arange(nTest):
    print("the classifier came back with: {0}, the real answer is: {1}" .format(int(a[t]), int(testLabel[t])))
    if (int(a[t])==int(testLabel[t])):
        accuracyTable[t]=1
    else:
        errors+=1
print("the total number of errors is: {}" .format(errors))
print("the total error rate is: {}" .format((errors/nTest)))
```

```python
print("훈련 세트 정확도: {:.2f}".format(svc.score(pca_train,trainLabel)))
print("테스트 세트 정확도: {:.2f}".format(svc.score(pca_test,testLabel)))
```

### 1.3.2. Applying k-NN
### Use the non-PCA version for k-NN

```python
import operator
```

```python
def classify0(inX, dataSet, labels, k):
    dataSetSize = dataSet.shape[0]
    diffMat = tile(inX, (dataSetSize,1))-dataSet
    sqDiffMat = diffMat**2
    sqDistances = sqDiffMat.sum(axis=1)
    distances = sqDistances**0.5
    sortedDistIndices = distances.argsort()
    classCount={}
    for i in range(k):
        votellabel = int(labels[sortedDistIndices[i]])
        classCount[votellabel] = classCount.get(votellabel,0)
    sortedClassCount = sorted(classCount.items(), key=operator.itemgetter(1),reverse=True)
    return sortedClassCount[0][0]
```

```python
errors = 0
accuracyTable=zeros(nTest)
for kn in arange(testData.shape[0]):
    a=classify0(testData[kn], trainData, trainLabel,3)
    print("the classifier came back with: {0}, the real answer is: {1}" .format(int(a), int(testLabel[kn])))
    if a==int(testLabel[kn]):
        accuracyTable[kn]=1
    else:
            errors+=1
print("the total number of errors is: {}" .format(errors))
print("the total error rate is: {}" .format((errors/nTest)))
```
